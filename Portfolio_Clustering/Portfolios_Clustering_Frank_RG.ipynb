```python
import pandas as pd
import numpy as np
df = pd.read_excel("Портфели_Frank.xlsx")
df = df.set_index('справочник')
df
```

```python
df.describe(include='all')
```

```python
df['Cash loans_share'] = df['Cash loans'] / df['Consumer loans']
df['Mortgage_share'] = df['Mortgage'] / df['Consumer loans']
df['Auto_share'] = df['Auto'] / df['Consumer loans']
df['CC&Overdrafts_share'] = df['CC&Overdrafts'] / df['Consumer loans']
df['POS_share'] = df['POS'] / df['Consumer loans']

df['Time deposits_share'] = df['Time deposits'] / df['Total Liabilities']
df['CASA_share'] = df['CASA'] / df['Total Liabilities']
df['Escrow_share'] = df['Escrow'] / df['Total Liabilities']
```
